```python
import              numpy as np
import             pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv("./MODIS_Ameriflux_input_data_final.csv")

print("Original df shape:", df.shape)

# Display the dataframe
df.head(10)
```

```python
# Read the CSV file
df1 = pd.read_excel("./Long_term_DI_CRU.xlsx")

print("Original df1 shape:", df1.shape)

# Display the dataframe
df1.head(10)
```

```python
# Calculate Ameriflux-based NPP using MODIS CUE and Ameriflux GPP, and store it in a new column
df["Ameriflux_NPP_C"] =           df["MODIS_CUE"] * df["Ameriflux_GPP"]

# Calculate Heterotrphic respiration by adding the computed Ameriflux-based NPP with Ameriflux_NEE
df["Ameriflux_Rh_C"]  =     df["Ameriflux_NPP_C"] + df["Ameriflux_NEE"]

df.head(10)
```

```python
# Replace empty strings with NaN so mean() can ignore them
df_clean = df.replace("", np.nan)

# Group by NEONSite and compute mean for ALL numeric columns (NaNs ignored)
df_neon_means = (
    df_clean
    .groupby("NEONSite", dropna=False, observed=False)
    .mean(numeric_only=True)
    .reset_index()
)

print("Averaged dataset shape:", df_neon_means.shape)
df_neon_means.head(26)
```

```python
# 1. Get the list of NEONSites present in df
sites_in_df = df["NEONSite"].unique()

# 2. Filter df1 to keep only those 26 sites
df1_matched = df1[df1["NEONSite"].isin(sites_in_df)].copy()

print("Original df1 NEONSite count:", df1["NEONSite"].nunique())
print("Matched NEONSite count:", df1_matched["NEONSite"].nunique())

df1_matched.head()
```

```python
cols_to_add = ["NEONSite", "Ameriflux_NEE", "Ameriflux_RE", "Ameriflux_GPP", "Ameriflux_NPP_C", "Ameriflux_Rh_C"]

df_combined = df1_matched.merge(
    df_neon_means[cols_to_add],
    on="NEONSite",
    how="left"
)

# Replace inf and -inf with NaN
df_combined = df_combined.replace([np.inf, -np.inf], np.nan)

df_combined_clean  = df_combined.dropna().reset_index(drop=True)
df_combined_clean.head(24)
```

```python
# Unique ecosystem classes
ecosystems_unique = df_combined_clean["Ecosystem"].unique()

# Assign each ecosystem a unique color
colors = plt.cm.tab10(np.linspace(0, 1, len(ecosystems_unique)))

# Map ecosystem  color
color_map = {ecosystems_unique[i]: colors[i] for i in range(len(ecosystems_unique))}

# Color for each point
point_colors = df_combined_clean["Ecosystem"].map(color_map)
```

```python
plt.figure(figsize=(8,4))

plt.scatter(
    df_combined_clean["DI"],
    df_combined_clean["Ameriflux_Rh_C"] / df_combined_clean["Ameriflux_NPP_C"],
    c=point_colors,
    label=None
)

plt.xlabel("DI")
plt.ylabel("Ameriflux_Rh_C / Ameriflux_NPP_C")
plt.tight_layout()

# Add legend manually
for eco, color in color_map.items():
    plt.scatter([], [], color=color, label=eco)

plt.legend(title="Ecosystem Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
